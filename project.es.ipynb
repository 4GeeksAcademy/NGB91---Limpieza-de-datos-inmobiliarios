#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

```python
import matplotlib.pyplot as plt
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

fuenlabrada = cinturon_sur[cinturon_sur['level5'] == 'Fuenlabrada']
variable_precios_fuenlabrada = (fuenlabrada['price'] - fuenlabrada['price'].min()) / (fuenlabrada['price'].max() - fuenlabrada['price'].min())
plt.hist(variable_precios_fuenlabrada, bins=30, color='indigo', edgecolor='black')


leganes = cinturon_sur[cinturon_sur['level5'] == 'Leganés']
variable_precios_leganes = (leganes['price'] - leganes['price'].min()) / (leganes['price'].max() - leganes['price'].min())
plt.hist(variable_precios_leganes, bins=30, color='darkblue', edgecolor='black')

getafe = cinturon_sur[cinturon_sur['level5'] == 'Getafe']
variable_precios_getafe = (getafe['price'] - getafe['price'].min()) / (getafe['price'].max() - getafe['price'].min())
plt.hist(variable_precios_getafe, bins=30, color='darkgreen', edgecolor='black')

alcorcon = cinturon_sur[cinturon_sur['level5'] == 'Alcorcón']
variable_precios_alcorcon = (alcorcon['price'] - alcorcon['price'].min()) / (alcorcon['price'].max() - alcorcon['price'].min())
plt.hist(variable_precios_alcorcon, bins=30, color='darkred', edgecolor='black')

plt.xlabel("Variable de Precios")
plt.ylabel("Frecuencia")
plt.title("Variable Precios Cinturón Sur")
plt.show()
```

Getafe es quien tiene la variable de precios más alta frente a Alcorcón que tiene la más baja. 
Alcorcón es el municipio con más variedad en su variable de precios. 
Leganés y Fuenlabrada son los que tienen más estática su variable de precios

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
import pandas as pd
from ipyleaflet import Map, basemaps, Marker, Popup
from ipywidgets import Label

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

coordenadas = []
for index, row in cinturon_sur.iterrows():
    poblacion = row['level5']
    latitud = row['latitude']
    longitud = row['longitude']
    coordenadas.append({
        'poblacion': poblacion,
        'latitud': latitud,
        'longitud': longitud
    })


map = Map(center=(40.35, -3.8), zoom=11.5, min_zoom=1, max_zoom=20, 
          basemap=basemaps.OpenStreetMap.Mapnik)


colores = {
    'Fuenlabrada': 'indigo',
    'Leganés': 'darkblue',
    'Getafe': 'darkgreen',
    'Alcorcón': 'darkred'
}

for coord in coordenadas: 
    poblacion = coord['poblacion']
    lat = float(coord['latitud'].replace(',', '.')) 
    lon = float(coord['longitud'].replace(',', '.'))

    marker = Marker(location=(lat, lon), draggable=False)
    marker.popup = Popup(child=Label(value=f"Población: {poblacion}"), close_button=True)

    map.add_layer(marker)

map
```
